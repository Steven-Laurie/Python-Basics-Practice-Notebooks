### Positive and Negative Indexing

```python
# You can reference an item in a list using an index, in the below example index number 2 is referenced. 
# All lists start at index 0 so when trying to reference the 1st item in the list you will use 0 and addition 1 for every item above the previous.
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[2])
```

```python
# You can also use negative indexing to access items in the reverse order, if we are trying to print the "white" colour we will reference index -2
# If you search for index "1" you will get the 2nd in order, if you search for index "-1" you will get the index in the last position. Position "0" in the index will always be the same. 
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[-2])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[:3])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[3:])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[:-3])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[-3:])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[-4:-2:])
```

```python
colours = ["red", "green", "blue", "yellow", "white", "black"]
print(colours[1:4])
```

### Updating Lists

```python
thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
# You can update an item in the list with a new name at a specific index using ["index number"]="new element name"

thislist[1]="blackcurrant"
print(thislist)
```

```python
# You can add an item to a list by using .append("name of element")

print(thislist)
thislist.append("orange")
print(thislist)
```

```python
# You can insert a new element at a specific index location

print(thislist)

thislist.insert(4,"watermelon")
print(thislist)
```

```python
# You can reverse a list using reverse() method

thislist.reverse()
```

```python
# You can use the pop() method to remove an element at a specified location

thislist.pop(2)
print(thislist)
```

```python
# You can also use del() and use [] within brackets to specify an index number to delete

del(thislist[2])
print(thislist)
```

```python
# You can sort the list using .sort

thislist.sort()
print(thislist)
```

```python
# And sort it in reverse order using .sort(reverse=True)

thislist.sort(reverse=True)
print(thislist)
```
